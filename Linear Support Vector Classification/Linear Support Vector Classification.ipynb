### Import Packages and Modules

```python
import pandas as pd
import pylab as pl
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import jaccard_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve

from sklearn.svm import LinearSVC
```

### Avoid Unwanted Warnings

```python
import warnings
warnings.filterwarnings("ignore")
```

### Load Dataset

```python
df = pd.read_csv("ChurnData.csv", na_values="?")
df.head(10)
```

### Preliminary Deal with Dataset

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.replace("?", np.NaN)
df.head(10)
```

```python
df.fillna(round(df.mean(),2), inplace=True)
df.head(10)
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
df.info()
```

```python
df['churn'] = df['churn'].astype('int')
df['confer'] = df['confer'].astype('float')
```

```python
df.info()
```

### Define Independent and Dependent Variables for Implement Models

```python
X = np.asarray(df[['tenure', 'age', 'address', 'income', 'ed', 'employ', 'equip', 'callcard', 'wireless','longmon', 'tollmon', 'equipmon', 'cardmon', 'wiremon', 'longten', 'tollten', 'cardten', 'voice', 'pager', 'internet', 'callwait', 'confer', 'ebill', 'loglong', 'logtoll', 'lninc', 'custcat']])
y = np.asarray(df['churn'])
```

```python
X[0:5]
```

```python
y[0:5]
```

### Normalize Dataset

```python
X = preprocessing.StandardScaler().fit(X).transform(X)
X[0:5]
```

### Spliting for Train & Test Dataset

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.25, random_state=4)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

### Linear Support Vector Classification

```python
linear_svc = LinearSVC()
linear_svc.fit(X_train, y_train)
Y_pred_lsvc = linear_svc.predict(X_test)
acc_linear_svc = round(linear_svc.score(X_test,y_test) * 100, 2)
print("Linear Support Vector Classification Accuracy: ", acc_linear_svc)
```

```python
cvs_lsvc = round((cross_val_score(linear_svc, X,y,cv=10,scoring='accuracy')).mean()*100,2)
print('Cross Validated Score:', cvs_lsvc)
```

```python
class_report = classification_report(y_test, Y_pred_lsvc)
print("Classificiation Report: \n", class_report)
```

### Confusion Matrix

```python
cnf_matrix = confusion_matrix(y_test, Y_pred_lsvc, labels=[1,0])
sns.heatmap(cnf_matrix,annot=True,fmt='3.0f',cmap="Greens")
plt.title('Confusion Matrix (Linear Support Vector Classification)', y=1.05, size=15)
```

### Jaccard Score

```python
jac_score_linear_svc = round(jaccard_score(y_test, Y_pred_lsvc, pos_label=0) * 100, 2)
print("Jaccard Score: ", jac_score_linear_svc)
```

### Mean Absolute Error & Mean Squard Error

```python
lsvc_mae = round((mean_absolute_error(y_test, Y_pred_lsvc)*100), 2)
lsvc_mse = round((mean_squared_error(y_test, Y_pred_lsvc)*100), 2)
print("Mean Absolute Error: ", lsvc_mae)
print("Mean Squared Error: ", lsvc_mse)
```

```python
auc_lsvc = roc_auc_score(y_test, Y_pred_lsvc)
fpr_lsvc, tpr_lsvc, thresholds_lsvc = roc_curve(y_test, Y_pred_lsvc)
plt.figure(figsize=(12, 7))
plt.plot(fpr_lsvc, tpr_lsvc, label=f'AUC (Linear SVC) = {auc_lsvc:.2f}')
plt.title('ROC AUC Score')
plt.plot([0, 1], [0, 1], color='blue', linestyle='--', label='Baseline')
plt.legend()
```
